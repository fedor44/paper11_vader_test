```python
# 감정 분석을 위해 nltk의 vader 모델을 사용하기 위해 패키지 다운로드
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
nltk.download("all")
import pandas as pd

# 감정 분석 수행
sia = SentimentIntensityAnalyzer()
df_lexicon = pd.read_csv('/content/drive/MyDrive/DeepNeuralNetwork/lexiconData.csv')
# 두 열의 값을 더하고 평균 내어 각 어휘의 감정 점수를 계산
df_lexicon['sentiment'] = (df_lexicon['Aff_Score'] + df_lexicon['Neg_Score'])/2
# Item 열을 키(key)로, sentiment 열을 값(value)으로 갖는 딕셔너리로 변환
df_lexicon = dict(zip(df_lexicon.Item, df_lexicon.sentiment))
# 하나의 단어로 구성된 항목만을 선택하여 딕셔너리 업데이트
df_lexicon = {k:v for k,v in df_lexicon.items() if len(k.split(' '))==1}
temp_dict = {}

# value 조건에 value 값 재계산 후 딕셔너리에 추가
for k, v in df_lexicon.items():
  if v > 0:
    temp_dict[k] = v / max(df_lexicon.values()) * 4
  else:
    temp_dict[k] = v / min(df_lexicon.values()) * -4

final_dict = {}
# 딕셔너리에 주식 시장 관련 어휘와 감정 점수 업데이트
final_dict.update(temp_dict)
# 딕셔너리에 sia.lexicon의 어휘와 감정 점수 업데이트
final_dict.update(sia.lexicon)
# 하나의 딕셔너리로 결합된 정보를 sia.lexicon에 할당
# 주식 시장 관련 어휘와 기존의 감정 어휘를 모두 활용하여 감정 점수 계산 가능
sia.lexicon = final_dict
```

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

# 테스트 뉴스 헤드라인
"""
sentences = [
    "According to Gran , the company has no plans to move all production to Russia , although that is where the company is growing .",
    "For the last quarter of 2010 , Componenta 's net sales doubled to EUR131m from EUR76m for the same period a year earlier , while it moved to a zero pre-tax profit from a pre-tax loss of EUR7m .",
    "Sales in Finland decreased by 10.5 % in January , while sales outside Finland dropped by 17 % .",
    "Operating profit fell to EUR 35.4 mn from EUR 68.8 mn in 2007 , including vessel sales gain of EUR 12.3 mn .",
    "Both operating profit and turnover for the nine-month period increased , respectively from EUR2 .4 m and EUR43 .8 m , as compared to the corresponding period a year ago ."
]
"""
# financial_phrasebank =
# 1(neutral), 2(positive), 0(negative), 0(negative), 2(positive)

# 데이터셋 로드
data = pd.read_csv('/content/drive/MyDrive/paper/financial_phrasebank_auditor_sentiment.csv')

# 데이터셋 준비
sentences = data['sentence']
labels = data['label']

# 예측 레이블과 실제 레이블을 저장할 리스트 초기화
y_pred = []
y_true = labels

# 감성 점수 계산
def analyze_sentiment(text):
    # sia 객체는 이전 단계에서 이미 생성되었다고 가정
    score = sia.polarity_scores(text)

    # compound 점수를 사용하여 긍정, 부정, 중립 판단
    # 임계값 가장 잘 나온 값으로 설정
    # 보수적 분류: 0.65과 -0.65과 같은 높은 임계값은 모델이 매우 확신하는 경우에만 긍정적 또는 부정적으로 분류하도록 함
    # 이는 분류가 매우 보수적이 되며, 대부분의 텍스트가 중립으로 분류될 가능성이 높아짐
    # 실제로 긍정적이거나 부정적인 감정이 약간 포함된 텍스트조차 중립으로 분류될 수 있음

    # 일반적 관례에 따라 0.05, -0.05 로 셋팅
    # VADER가 개발될 당시부터 많이 사용되었으며, 대부분의 일반적인 케이스와 데이터셋에서 합리적으로 잘 작동하는 것으로 나타남
    compound_score = score['compound']
    if compound_score >= 0.05:
        return 2 #'Positive'
    elif compound_score <= -0.05:
        return 0 #'Negative'
    else:
        return 1 #'Neutral'

predictions = []
for sentence in sentences:
  prediction = analyze_sentiment(sentence)

  y_pred.append(prediction)

# 정밀도, 재현율, F1 점수 및 정확도 계산
precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred, average='weighted')
accuracy = accuracy_score(y_true, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Compute confusion matrix
cm = confusion_matrix(y_true, y_pred)

# Plot confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['Negative', 'Neutral', 'Positive'], yticklabels=['Negative', 'Neutral', 'Positive'])
plt.title('Confusion Matrix')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.show()
```
